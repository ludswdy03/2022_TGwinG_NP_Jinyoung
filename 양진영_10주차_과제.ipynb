* 구현해야할 클래스인 Members의 명세는 아래와 같다
* Linked List를 사용해서 풀이해야 함, Linked List에 대한 설명은 이 블럭의 마지막 부분 [참고 자료] 에 있음

### ```Class Members``` 
##### 멤버 변수
* ```head = Member('headData', 'headData', 'headData', 'headData', False)```
* ```number = 0```

##### 멤버 함수
* ```addMember(self, id, year, name, gender, registered)```
    * id: 학번 (str)
    * year: 가입년도 (str)
    * name: 이름 (str)
    * gender: 성별 (str)
    * registered: 회비 납부 여부 (boolean)
    <br><br>
* ```delMember(self, id)```
    * 입력 받은 학번과 일치하는 회원을 삭제한다
    * 일치하는 회원이 없거나 회원 수가 0 이라면 "can not find %s member" %id를 출력하고 함수를 종료한다
<br><br>
* ```editMember(self, id)```
    * 입력 받은 학번과 일치하는 회원의 정보를 수정하는 함수를 불러온다
    * 일치하는 회원이 없거나 회원 수가 0 이라면 "can not find %s member" %id를 출력하고 함수를 종료한다
<br><br>
* ```editMemberDetail(self)```
    * 회원의 정보를 출력하고, 다음 줄에 어떤 정보를 수정할지 물어보는 텍스트를 출력한 후 사용자로 부터 입력 받는다.
    * 대체할 정보를 입력 받은 후 저장한다
<br><br>
* ```isLen(self)```
    * 동아리의 회원 수를 반환한다
<br><br>
* ```newSeason(self)```
    * 새 학기를 맞아 회비 납부 여부를 초기화한다
    * 모든 회원의 회비 납부 여부를 False로 바꾼다
<br><br>
* ```infoMem(self, id)```
    * 입력 받은 학번과 일치하는 회원의 정보를 출력한다
<br><br>
* ```printMembers(self)```
    * 모든 회원의 정보를 아래와 같이 보기 좋게 출력한다

|id|year|name|gender|registered|
|:-:|:-:|:-:|:-:|:-:|

# 참고 자료
[강의노트 18. 자료구조 - LinkedList (링크드 리스트)](https://wayhome25.github.io/cs/2017/04/17/cs-19/) - 초보몽키의 개발공부로그

````python
class Member:
    id = None
    name = None
    year = None
    gender = None
    registered = None
    next = None

    def __init__(self, id, year, name, gender, registered):
        self.id = id
        self.name = name
        self.year = year
        self.gender = gender
        self.registered = registered
    def getInfo(self):
        return self.id, self.year, self.name, self.gender, self.registered
````

````python
class Members:
    head = Member('headData', 'headData', 'headData', 'headData', False)
    number = 0
    
    def addMember(self, id, year, name, gender, registered):
        member = Members.head
        number = Members.number
        while member.next:
            member = member.next
        member.next = Member(id, year, name, gender, registered)
        number = number + 1
        Members.number = number
    
    def delMember(self, id):
        member = Members.head.next
        number = Members.number
        prev_member = Members.head
        count = 0
        no_member = 0
        if number == 0:
            print("can't not find %s member" %id)
            no_member = no_member + 1
        while member:
            if id == member.id:
                prev = prev_member
                delete = prev.next
                prev.next = delete.next
                delete.next = None
                number = number - 1
                Members.number = number
                break
            else:
                count = count + 1
                prev_member = member
                member = member.next
        if count == number and no_member == 0:
            print("can't not find %s member" %id)
        
    def editMember(self, id):
        member = Members.head.next
        number = Members.number
        count = 0
        while member:
            if id == member.id:
                Members.editMemberDetail(member)
                break
            else:
                count = count + 1
                member = member.next
        if count == number:
            print("can't not find %s member" %id)
        
    def editMemberDetail(self):
        member = self
        change = input("어떤 정보를 바꾸시겠습니까?: ")

        if change == "year":
            year = input("수정할 가입년도 입력: ")
            member.year = year
        elif change == "name":
            name = input("수정할 이름 입력: ")
            member.name = name
        elif change == "gender":
            gender = input("수정할 성별 입력: ")
            gender.name = gender
        elif change == "registered":
            registered = input("수정할 회비 납부여부 입력: ")
            member.registered = registered

        
    def isLen(self):
        number = Members.number
        return number
        
    def newSeason(self):
        member = Members.head.next
        while member:
            member.registered = False
            member = member.next
        
    def infoMem(self, id):
        member = Members.head.next
        while member:
            if id == member.id:
                print("|     id    ",  "|    year    |", "   name    |", "  gender  |", "registered |")
                print("|", member.id, "|   ", member.year, "   |  ", member.name, "  |    ",
                 member.gender, "   |   ", member.registered, "   |")
                break
            else:
                member = member.next
        
    def printMembers(self):
        member = Members.head.next
        print("|     id    ",  "|    year    |", "   name    |", "  gender  |", "registered |")
        while member:
            print("|", member.id, "|   ", member.year, "   |  ", member.name, "  |    ",
                 member.gender, "   |   ", member.registered, "   |")
            member = member.next
````
